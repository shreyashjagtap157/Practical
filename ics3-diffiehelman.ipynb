```python
### PUBLIC KEYS
p = 19
def get_primitive_root(p):
    flag=False
    for q in range(1,p):
        values={}
        
        for x in range(0,p-1):
            val=pow(q,x,p)
            
            if val in values.keys():
                break
                
            values[val]=True
            
            if x==p-2:
                flag=True
        if flag==True:
            return q
    return None
q=get_primitive_root(p)
print(q)
```

```python
### PRIVATE KEYS
a = 13
b = 9
```

```python
### GENERATE KEYS
x = (q ** a) % p
y = (q ** b) % p
print('Generated Keys - x: {',x,'}, y: {',y,'}')
```

```python
### GENERATE SECRET KEY WITH x AND y
key_a = (y ** a) % p
key_b = (x ** b) % p
print(f'Generated Secret Keys - KEY-A: {key_a}, KEY-B: {key_b}')
```
