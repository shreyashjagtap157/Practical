```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.array([10, 9, 2, 15, 10, 16, 11, 16])
y = np.array([95, 80, 10, 50, 45, 98, 38, 93])
```

```python
plt.xlabel('No of hours spent driving')
plt.ylabel('Risk Score')
plt.scatter(x,y,color='black',marker='*')
plt.title('linear regression')
```

```python
def getCoef(x,y):
    #mean of x and y
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    
    #Total no of values
    n = len(x)
    
    # formula to calculate b1 and b2
    numer = 0
    denom = 0
    for i in range(n):
        numer += (x[i] - mean_x) * (y[i] - mean_y)
        denom += (x[i] - mean_x) ** 2
    b1 = numer / denom
    b0 = mean_y - (b1 * mean_x)
    
    return(b0, b1)
```

```python
coefs_ = getCoef(x,y)
print("b0 value =",coefs_[0])
print("b1 value =",coefs_[1])
```

```python
plt.xlabel('No of hours spent driving')
plt.ylabel('Risk Score')
plt.title('linear regression')
plt.scatter(x,y,color='black',marker='*')
y_pred = coefs_[0] + coefs_[1]*x
plt.plot(x, y_pred, color = "r")
```

```python
from math import sqrt
sum_err = 0.0
for i in range(len(y)):
    pred_err = y_pred[i] - y[i]
    sum_err += pred_err**2
    mean_err = sum_err / float(len(y))
sqrt(mean_err)
```
